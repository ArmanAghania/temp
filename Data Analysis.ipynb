# Preprocessing

```python
# Modify the persian_to_gregorian function to handle integer dates
def persian_to_gregorian(persian_date):
    # Define the months and their number of days
    persian_months = [
        ('Farvardin', 31),
        ('Ordibehesht', 31),
        ('Khordad', 31),
        ('Tir', 31),
        ('Mordad', 31),
        ('Shahrivar', 31),
        ('Mehr', 30),
        ('Aban', 30),
        ('Azar', 30),
        ('Dey', 30),
        ('Bahman', 30),
        ('Esfand', 29)
    ]
    # Parse the input date
    persian_date = str(persian_date)
    year = int(persian_date[:4])
    month = int(persian_date[4:6])
    day = int(persian_date[6:])
    # Check if the year is a leap year
    is_leap = year % 4 == 3
    if is_leap:
        persian_months[11] = ('Esfand', 30)
    # Calculate the day of the year
    day_of_year = sum(days for month_name, days in persian_months[:month-1]) + day
    # Calculate the Gregorian date
    gregorian_date = datetime.date(year + 621, 3, 21) + datetime.timedelta(days=day_of_year-1)
    return gregorian_date

# Test the function
persian_to_gregorian(14020608)
```

```python
# Drop the 'DateStartStop', 'TimeStartStop', 'DateEndStop', and 'TimeEndStop' columns
data_cleaned = joined_data.drop(columns=['Malfunction Start Date', 'Malfunction Start Time', 'Malfunction End Date', 'Malfunction End Time'])

# Drop rows with null values in 'Request Date' column
data_cleaned.dropna(subset=['Request Date'], inplace=True)

# Convert 'Request Date' column to integers
data_cleaned['Request Date'] = data_cleaned['Request Date'].astype('int')

# Convert the 'Request Date' column to Gregorian dates
data_cleaned['Request Date'] = data_cleaned['Request Date'].apply(persian_to_gregorian)

# Set the 'تاریخ درخواست' column as the index of the dataframe
data_cleaned.set_index('Request Date', inplace=True)

data_cleaned.head()
```

# Predictive Analysis

```python
# Split the data into a training set and a test set
train = daily_malfunctions[:-30]
test = daily_malfunctions[-30:]

# Plot the training and test sets
plt.figure(figsize=(12, 6))
plt.plot(train.index, train.values, label='Train', color='b')
plt.plot(test.index, test.values, label='Test', color='r')
plt.title('Number of Malfunctions Reported Each Day')
plt.xlabel('Date')
plt.ylabel('Number of Malfunctions')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot ACF and PACF to determine the order of the AR and MA components
fig, ax = plt.subplots(1, 2, figsize=(12, 4))
plot_acf(train, ax=ax[0])
plot_pacf(train, ax=ax[1])
plt.show()

# Fit the SARIMA model
model = SARIMAX(train, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
fit = model.fit(disp=False)

# Print the summary of the model
print(fit.summary())
```

```python
# Forecast the number of malfunctions for the next 30 days
forecast = fit.get_forecast(steps=30).predicted_mean

# Plot the actual and forecasted values
plt.figure(figsize=(12, 6))
plt.plot(train.index, train.values, label='Train', color='b')
plt.plot(test.index, test.values, label='Test', color='r')
plt.plot(forecast.index, forecast.values, label='Forecast', color='g')
plt.title('Number of Malfunctions Reported Each Day')
plt.xlabel('Date')
plt.ylabel('Number of Malfunctions')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split

# Create features based on the date
X = pd.DataFrame(index=daily_malfunctions.index)
X['day_of_week'] = X.index.dayofweek
X['month'] = X.index.month

# Use the daily_malfunctions data as the target variable
y = daily_malfunctions

# Split the data into a training set and a test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

X_train.head()
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Calculate the mean squared error
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
print('Mean Squared Error:', mse)

# Plot the actual and predicted values
plt.figure(figsize=(12, 6))
plt.plot(y_test.index, y_test.values, label='Actual', color='b')
plt.plot(y_test.index, y_pred, label='Predicted', color='r')
plt.title('Number of Malfunctions Reported Each Day')
plt.xlabel('Date')
plt.ylabel('Number of Malfunctions')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Fit the Gradient Boosting model
from sklearn.ensemble import GradientBoostingRegressor
model = GradientBoostingRegressor(random_state=42)
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Calculate the mean squared error
mse = mean_squared_error(y_test, y_pred)
print('Mean Squared Error:', mse)

# Plot the actual and predicted values
plt.figure(figsize=(12, 6))
plt.plot(y_test.index, y_test.values, label='Actual', color='b')
plt.plot(y_test.index, y_pred, label='Predicted', color='r')
plt.title('Number of Malfunctions Reported Each Day')
plt.xlabel('Date')
plt.ylabel('Number of Malfunctions')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Import the XGBoost library
import xgboost as xgb

# Create a DMatrix object for the training and test sets
dtrain = xgb.DMatrix(X_train, label=y_train)
dtest = xgb.DMatrix(X_test, label=y_test)

# Set the parameters for the XGBoost model
param = {
    'max_depth': 3,
    'eta': 0.3,
    'objective': 'reg:squarederror',
    'eval_metric': 'rmse',
}
num_round = 100

# Train the XGBoost model
bst = xgb.train(param, dtrain, num_round)

# Make predictions on the test set
y_pred = bst.predict(dtest)

# Calculate the mean squared error
mse = mean_squared_error(y_test, y_pred)
print('Mean Squared Error:', mse)

# Plot the actual and predicted values
plt.figure(figsize=(12, 6))
plt.plot(y_test.index, y_test.values, label='Actual', color='b')
plt.plot(y_test.index, y_pred, label='Predicted', color='r')
plt.title('Number of Malfunctions Reported Each Day')
plt.xlabel('Date')
plt.ylabel('Number of Malfunctions')
plt.legend()
plt.grid(True)
plt.show()
```
